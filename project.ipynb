```python
import numpy as np
import pandas as pd
import zipfile
from matplotlib import pyplot as plt

#insert the name of the archive here
source_file = "dataset.zip"
#use the appropiate py filetype to access the contents
zipped_data = zipfile.ZipFile(source_file)
#get the names of the files in the archive(zip)
zd_info = zipped_data.namelist()
#separate the delimiter from the useful id data
zd_aux = [x.split("animal_data/")[1] for x in zd_info]
#labels for the images will be set as the folder name in which the images are placed in
labels = [x.split("/")[0] for x in zd_aux]
#image id will be made from the name of the files (they are all uniquely named)
ids = [x.split("/")[1]for x in zd_aux]
#create a dictionary out of them
data = {"imageID" : ids, "label" : labels}
#convert the dictionary to a dataframe
df = pd.DataFrame.from_dict(data)
#check if the data is imported corectly by checking data distribution
vc = df["label"].value_counts()
#get the labels and associated counts
vals = vc.values
indx = vc.index 
#plot them
plt.title("Data distribution")
plt.xlabel("Count")
plt.ylabel("Label")
plt.barh(indx,vals)
plt.show()
```
